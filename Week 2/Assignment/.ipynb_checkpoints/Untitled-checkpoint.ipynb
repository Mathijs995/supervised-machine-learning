```python
res = linear_model.ElasticNetCV(
  l1_ratio = params_list['alpha'],
  alphas = params_list['lambda'],
  fit_intercept = True,
  normalize = False,
  max_iter = 1000,
  tol = 10e-6,
  cv = 5,
  verbose = 0,
  n_jobs = -1,
).fit(X, y)
```
